```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fifa = pd.read_csv('fifa_data.csv')
labels = ['Left', 'Right']
colors = ['yellow', 'blue']
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]


plt.pie([left,right], labels=['Left', 'Right'], colors=['blue', 'green'], autopct='%.2f %%' ,
       radius=1.2 )


plt.title('Foot Preference of Fifa Players')

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('Solarize_Light2')

a1 = fifa.loc[fifa['Age'] < 20].count()[0]
a2 = fifa.loc[(fifa['Age'] >= 20) & (fifa['Age'] < 25)].count()[0] 
a3 = fifa.loc[(fifa['Age'] >= 25) & (fifa['Age'] < 30)].count()[0]
a4 = fifa.loc[(fifa['Age'] >= 30) & (fifa['Age'] < 35)].count()[0]
a5 = fifa.loc[fifa['Age'] >= 35].count()[0]


ages = [a1,a2,a3,a4,a5]
labels = ['under 20','20-25', '25-30', '30-35', 'over 35']
explode = (.4,0,0,0.1,.4)

plt.pie(ages, labels= labels, autopct='%.2f %%', explode=explode) #pcdistance= 0.8 percent distance
plt.title('Age of Professional Soccer Players in 2018 by Mirsaid', fontdict={'fontweight': 'bold', 'fontsize' : 12})

plt.show()


```
