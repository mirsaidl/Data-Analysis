```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]


# Resize your graph
# Dpi is pixel in per inch 
plt.figure(figsize=(5,3) , dpi=500)

#Line 1


# Shorthand notation
# format -> '[color][marker][line]

plt.plot(x,y, 'g*--', label='2x')
# b is color * is marker point -- line

# Line 2

# select interval we want to lot points at
x2 = np.arange(0, 4.5,0.5)
print(x2)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r' , label = 'X^2')

# Plot reminder of graph
plt.plot(x2[5:], x2[5:]**2, 'r.--' )

# Keyword Argument Notation
# plt.plot(x,y, label = "1x", color = 'red', linewidth = 2, marker = '.', markersize = 10)

#Add a title (font parameters with fontdict)
plt.title("Our First Graph", fontdict={"fontname" : "Comic Sans MS", 'fontsize' : 20}) # title

# X and L labels
plt.xlabel("X Axis!")
plt.ylabel("Y Axis!")

# X,Y tickmarks (scale or parameters of graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,7,10])

 
plt.legend() # for use of label

#Save
plt.savefig('mygraph.png', dpi =500)

#Show plot
plt.show() # erases unimportant parts

```
