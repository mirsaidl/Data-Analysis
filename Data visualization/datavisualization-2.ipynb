```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('/kaggle/input/tashkent-real-estate-2019/uybor.xlsx')
df.head()
```

```python
sns.barplot(x=df.district ,y=df.price, color='red')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x=df.rooms ,y=df.price, color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, x="rooms")
plt.show()
```

```python
sns.violinplot(data=df, x="district", y="price")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(data=df, diag_kind='kde')
```

```python
sns.jointplot(data=df, x="price", y="size", kind="scatter")
```
