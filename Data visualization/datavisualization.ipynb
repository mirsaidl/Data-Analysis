```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x = np.arange(0, 361, 30)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, sin, color='#9A6CBE', marker='o', linewidth=2.5, linestyle='--', alpha=0.7)
plt.plot(x, cos, color='g', marker='o', linewidth=2.5, linestyle=':', alpha=0.7)

plt.title('Gradius sin and cos')

tics = [0,30,60,90,120,150,180,210,240,270,300,330,360]
plt.xticks(tics)

plt.grid(color='k', alpha=0.5, linestyle='--')
plt.legend(['sin', 'cos'], loc=0, fontsize='x-small')

plt.show()
```

# Distribution

```python
df=df[df.year>=2010]
df=df[df.transmission!='Other']
df=df[df.fuelType!='Other']
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.xticks(rotation=90)
plt.show()
```

```python
# plt.figure(figsize=(10,5))
sns.displot(data=df, x='price', hue='transmission', multiple='stack', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()

# https://jovian.com/outlink?url=https%3A%2F%2Fseaborn.pydata.org%2Fgenerated%2Fseaborn.displot.html
```

```python
# plt.figure(figsize=(10,5))
sns.displot(data=df, x='price', col='transmission', multiple='stack', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
# KDE
#displot
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.kdeplot(data=df, x='price', hue='transmission',shade=True)
plt.xticks(rotation=90)
plt.show()
```

```python
# plt.figure(figsize=(10,5))
sns.displot(data=df, x='price', hue='transmission', multiple='stack', kde=True, height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", y="year", kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
# Ecdf
sns.displot(data=df, x="price", hue="fuelType" ,kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

# SubPlots

```python
plt.figure(figsize=(10,5))

# sin
plt.subplot(2,1,1)
plt.plot(x, sin, color='#9A6CBE', marker='o', linewidth=2.5, linestyle='--', alpha=0.7)
tics = [0,30,60,90,120,150,180,210,240,270,300,330,360]
plt.xticks(tics)
plt.grid(color='k', alpha=0.5, linestyle='--')

# cos
plt.subplot(2,1,2)
plt.plot(x, cos, color='g', marker='o', linewidth=2.5, linestyle=':', alpha=0.7)
tics = [0,30,60,90,120,150,180,210,240,270,300,330,360]
plt.xticks(tics)
plt.grid(color='k', alpha=0.5, linestyle='--')


plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(20,10))

# ax[0]
sns.histplot(ax=ax[0], data=df, x='price', hue='transmission', multiple='stack', kde=True)
plt.xticks(rotation=90)

# ax[1]
sns.scatterplot(ax= ax[1], data=df, x='mileage', y='price' , hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,10))

fig.suptitle('CARS')

# ax[0,0]
sns.histplot(ax=ax[0,0], data=df, x='price', hue='transmission', multiple='stack', kde=True)
# ax[0,1]
sns.scatterplot(ax=ax[0,1], data=df, x='mileage', y='price' , hue='transmission', style='fuelType', s=100)
# ax[1,0]
sns.scatterplot(ax=ax[1,0], data=df, x='year', y='price' , hue='transmission', style='fuelType', s=100)
# ax[1,1]
sns.histplot(ax=ax[1,1], data=df, x='mileage', hue='fuelType')

ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")

plt.show()
```
