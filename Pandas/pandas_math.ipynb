**Function Pandas**

```python
data = {
    'Population': [331, 1440, 1393, 213, 146, 126, 83, 68],
    'GDP': [22.675, 16.227, 2.875, 1.449, 1.708, 6.266, 4.339, 2.829],
    'Life Expectancy': [78.8, 76.7, 69.7, 75.9, 72.6, 84.6, 80.9, 81.4],
    'Happiness Index': [7.2, 5.2, 4.0, 6.7, 5.5, 6.0, 7.6, 7.4],
    'Internet Users': [312, 989, 624, 149, 109, 118, 77, 64],
    'Number of Nobel Laureates': [383, 10, 10, 4, 28, 28, 83, 130]
}
df = pd.DataFrame(data)
df.index = ['USA', 'China', 'India', 'Brazil', 'Russia', 'Japan', 'Germany', 'UK']
df
```

```python
np.round(df) # absolute
```

```python
df.mean()
```

```python
df.apply(np.mean)
```

```python
df.apply(np.max)
```

```python
df.apply(np.sum, axis='columns')
```

```python
f = lambda x: x.max() - x.min()
df.apply(f)
```

```python
def foo(x):
    return pd.Series([x.min(), x.max(), x.mean()], index=['Min', 'Max', 'Mean'])
df.apply(foo)
```

```python
df
```

```python
def gdp_per_capita(x):
  return (x.GDP * 1e12)/(x.Population * 1_000_000) # 1e12 = 1 trillion
# We found GDP per capita (by exact amount)
gdp_pc = df.apply(gdp_per_capita, axis=1)
```

```python
np.round(gdp_pc)
```
