```python
import numpy as np
import pandas as pd
```

# Dropna()

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col='index')
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()/len(df) * 100
```

```python
# DROPNA
df.dropna()
```

```python
df.dropna().shape # too small data
```

```python
df.dropna(axis=1)
```

```python
df.dropna(subset=['wheel-base', 'length', 'horsepower'])
```

```python
df.dropna(axis=1,thresh=40)
```

```python
df.dropna(subset = ['price', 'horsepower'])
```

# Fillna()

```python
df.head()
```

```python
df.fillna(0) # Nan = 0
```

```python
df.fillna(df.mean())
```

```python
df.fillna({'wheel-base' : 90, 'horsepower' : 100})
```

```python
df[df.company == 'audi'].fillna(method='ffill', inplace=True) # forward fill
df
```

```python
df[df.company == 'toyota'].fillna(method='bfill') # back fill
```

```python
df.fillna({'horsepower':np.mean(df['horsepower']), "wheel-base":np.mean(df['wheel-base'])})
```

# Map()


```python
df.head()
```

```python
df['num-of-cylinders'].unique()
```

```python
dict = {'four': 4, 'five': 5, 'six': 6, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8 }
df['cylinders'] = df['num-of-cylinders'].map(dict)
df.drop("num-of-cylinders",axis=1, inplace=True)
df
```

```python
df.index = range(1,62)
df
```

```python
# lets turn price from usd to uzs
df['price'] = df.price.map(lambda x: x*12336)
df
```
