```python
import numpy as np
import pandas as pd
```

# Cut, Qcut

```python
df = pd.read_csv('university.csv',usecols = ['university','rank', 'overall_score', 'employment_outcomes'], index_col='university')
df.head()
```

```python
df.shape
```

```python
df.overall_score.describe()
```

```python
bins = [50,70,80,90,100]
score = pd.cut(df.overall_score,bins)
score
```

```python
score.value_counts()
```

```python
df['ovscore_bins'] = score
df.sample(7)
```

```python
df.employment_outcomes.describe()
```

```python
bins = [10,30,50,70,90,100]
emp = pd.cut(df.employment_outcomes,bins)
emp
```

```python
df['emp_range'] = emp
df
```

```python
name = ['bad', 'average', 'good', 'excelent', 'prestigious' ]
emp = pd.cut(df.employment_outcomes,bins, labels=name)
emp.value_counts()
```

```python
df['emp_range'] = emp
df
```

```python
df[df.emp_range == 'prestigious']
```

```python
emp = pd.cut(df.employment_outcomes, 5)
emp.value_counts()
```

```python
 # if equally divided
emp = pd.qcut(df.employment_outcomes, 5)
emp.value_counts()
```

# groupby

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.groupby('company').mean()
```

```python
df.groupby('company').price.max()
```

```python
df.groupby(['company', 'body-style']).mean()
```

```python
df.groupby('company').max()
```

```python
df.groupby('company').agg([len,min,max])
```

```python
df.groupby('company').price.agg([min,max,np.mean])
```

```python
df
```

```python
df.price.describe()
```

```python
bins = [5000,10000,20000,30000,40000]
labels = ['cheap', 'average', 'expensive', 'costly']
df['price_labels'] = pd.cut(df.price, bins, labels=labels)
df
```
