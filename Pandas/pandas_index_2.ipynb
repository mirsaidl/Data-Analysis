**loc/iloc vs at/iat**

```python
df
```

```python
# loc (call by name)
df.loc[['2015 Population', '2020 Population'],['China','India','USA']]
```

```python
# iloc (call by index number)
df.iloc[[0,2],[0,1,2,4,6]]
```

```python
df.at['2020 GDP','Japan']
```

```python
# at (call by name) (exact thing)
df.at['2020 GDP','Japan'] - df.at['2015 GDP','Japan']
```

```python
# iat (call by index) (exact thing)
df.iat[2,6] # Japan 2020 GDP
```

```python
df.at['2020 GDP','Japan'] == df.iat[2,6] # true
```
