# Web

```python
import pandas as pd
import numpy as np
import requests, json
```

```python
bukhara = pd.read_html('https://en.wikipedia.org/wiki/Bukhara')
print(len(bukhara))
df = bukhara[1]
```

```python
df = df.T
df
```

```python
df.drop(6,axis=1,inplace=True)
df.drop(7,axis=1,inplace=True)
```

```python
df.columns
df.set_index(('Climate data for Bukhara (1991–2020)', 'Month'))
```

```python
df.index.name = 'Facts/Months'
df
```

```python
df.to_excel('bukhara.xlsx')
```
