```python
import numpy as np
import pandas as pd
import sqlite3 as sql
```

***# Creating, Writing, Editing***

```python
db = sql.connect('world_info')
cursor = db.cursor()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df
```

```python
cols = ['entity', 'code', 'year', 'coal', 'gas', 'hydro','other', 'solar', 'oil', 'wind', 'nuclear']
df.columns = cols
df
```

```python
df.to_sql('electr', db)
```

```python
query = "SELECT * FROM electr"
cursor.execute(query)
df2 = pd.read_sql(query,db, index_col="index")
```

# Open new table

```python
query = """CREATE TABLE IF NOT EXISTS population(
rank INT,
country TEXT PRIMARY KEY,
density DOUBLE,
population DOUBLE NOT NULL,
area INT NOT NULL
)"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info(population);")
cursor.fetchall()
```

```python
query = "SELECT * FROM population"
df3 = pd.read_sql(query,db)
df3
```

```python
q = """ CREATE TABLE IF NOT EXISTS cars(
  rank INT,
  brand TEXT NOT NULL,
  price INT,
  model TEXT PRIMARY KEY,
  year INT,
  mileage INT NOT NULL
)
"""
cursor.execute(q)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```
