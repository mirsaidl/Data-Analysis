```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
from urllib.request import urlretrieve
url = "https://github.com/anvarnarz/praktikum_datasets/raw/main/electricity.db"
urlretrieve(url,'electricity.db')
```

```python
db = sql.connect('electricity.db')
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

```python
query  = "PRAGMA table_info('elect')"
cursor.execute(query)
cursor.fetchall()
```

```python
query = "SELECT * FROM elect LIMIT 10"
df = pd.read_sql(query,db)
df.sample(5)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan'"
dfuz = pd.read_sql(query,db)
dfuz.shape
```

```python
# DISTINCT
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query,db)
countries.head()
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' and year BETWEEN 2010 AND 2020"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
# BETWEEN
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country='Uzbekistan' and year BETWEEN 2010 AND 2020"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
# IN
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country='Uzbekistan' and year IN (2005,2010,2015,2020)"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
# LIKE
query = "SELECT * FROM elect WHERE country LIKE  '%tan' AND year IN (2015,2018)"
df = pd.read_sql(query,db)
df
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year=2020"
pd.read_sql(query, db)
```
