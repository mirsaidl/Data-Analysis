```python
db = sql.connect("chinook.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute('SELECT COUNT (*) FROM genres')
cursor.fetchone()
```

```python
cursor.execute('PRAGMA table_info(customers);')
cursor.fetchall()
```

```python
query = 'SELECT * FROM customers LIMIT 10'
df = pd.read_sql(query,db,index_col='CustomerId')
df
```

```python
query = 'SELECT * FROM customers ORDER BY COUNTRY'
df = pd.read_sql(query,db)
df
```

```python
query = 'SELECT * FROM customers WHERE country="USA" OR country="Italy"'
df = pd.read_sql(query,db)
df
```

```python
query = "SELECT * FROM customers ORDER BY Country DESC" # DECS = descending
df = pd.read_sql(query, db, index_col="CustomerId")
df.head(10)
```

```python
query = 'SELECT DISTINCT country from Customers'
cursor.execute(query)
print(cursor.fetchall())
```

```python
countries = pd.read_sql(query, db)
countries.head()
```
